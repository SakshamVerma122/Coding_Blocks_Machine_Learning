### Scatter Plots

```python
import numpy as np
from matplotlib import pyplot as plt

prices = np.array([1,2,3,4])**2
print(prices)

# In such a cases your array's value is plotted taken on the y_axis and x_axis = indexes of your array 
plt.plot(prices,"ko")
plt.plot(prices,"--",color="red")

plt.xlabel("indexes of array")
plt.ylabel("Array values given by us")

plt.title("Auto x_axis")
```

### Scatter plots

```python
# Adjusting size of any plot
plt.figure(figsize=(10,10))

# You have to declare the function first before all the
plt.style.use('seaborn')

x = np.sort(np.random.randint(-10,10,300))

# Parabola
y1 = x**2

# Line
y2 = 2*x+1

plt.scatter(x,y1,color="black",label = "jamun", marker = "^")
plt.scatter(x,y2,color="orange",label = "orange",marker = "*")
plt.legend()

plt.show()
```

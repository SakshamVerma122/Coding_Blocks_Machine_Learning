### Line Plot
![](https://www.geogebra.org/resource/s3jEhYHw/r4VLJVTBWp28wTuq/material-s3jEhYHw.png)

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# plt.plot?
```

```python
a = np.sort(np.random.randint(-10,10,300))

# Parabola
b1 = a**2

# Line
b2 = 2*a+1
```

```python
# It excludes the last element
k = np.arange(-10,10)
print(k)
```

```python
plt.plot(a,b1,color="red",marker="o")

# Used to indicate that one graph has been finished 
plt.show()


plt.plot(a,b2,color="green",linestyle="dashed")

# plt.show() is used to show the graphs
plt.show()
```

```python
plt.plot(a,b1,color="red",label ="Apple")

plt.plot(a,b2,color="green",label="Kiwi")

# For labeling x axis
plt.xlabel("Time")

# For labeling y axis
plt.ylabel("Price")

# Giving title to the graph
plt.title("Price of fruits over time")

plt.legend()

# plt.show() is used to show the graphs
plt.show()
```

* Let's take a closer look at how the for loop is actually implemented in Python.


internally, the for loop creates an iterator object, iter_obj by calling iter() on the iterable.

Ironically, this for loop is actually an infinite while loop.

Inside the loop, it calls next() to get the next element and executes the body of the for loop with this value. After all the items exhaust, StopIteration is raised which is internally caught and the loop ends. Note that any other kind of exception will pass through.

### Creating our own range function

```python
# It's both itterable and iterate
class yrange:
    # n is the number upto which I want the range
    def __init__(self,n):
        
        self.i = 0
        # itterator = i -> represents uptill which I have itterated
        
        self.n = n
        
    # This method makes our class iterable
    def __iter__(self):
        #It returns the itterator
        
        # here self.i is the itterator and yrange is it's enclosing class
        # In this case yrange is both our itterator and itterable
        # as it contains i(itterator) and as only i is needed to be printed till the range n
        # i.e why we are returning self as __itter__ returns itterate
        return self
    
    # this method should be implemented by the itterartor
    def __next__(self):
        if self.i < self.n:
            i = self.i
            self.i +=1
            return i
        else:
            raise StopIteration()
        
```

```python
for x in yrange(5):
    print(x)
```

```python
class zrange:
    def __init__(self,range_iterable,iterable):
        self.range_iterable = range_iterable
        self.iterable = iterable
        self.itterator = 0
        
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.itterator < self.range_iterable:
            itterator = self.iterable[self.itterator]
            self.itterator +=1
            return itterator
        else:
            raise StopIteration()
```

```python
for y in zrange(3,["Saksham","Shubham","Abhinav"]):
    print(y)
```

```python
# itterable
class zrange:
    
    def __init__(self,n):
        self.n = n
    
    def __iter__(self):
        return zrange_iter(self.n)
        # The reason why it will not return an empty list is that it's returning 
        # object of zrange_iter() and not self zrange_iter() object differs each time
        # but self remains same
        

# itterator
class zrange_iter:
    
    def __init__(self,n):
        self.i = 0
        self.n = n
        
    def __iter__(self):
        return self
    
    def __next__(self):
        
        if self.i < self.n:
            itterator = self.i
            self.i +=1
            return itterator
        
        else:
            raise StopIteration()
        
```

```python
for x in zrange(10):
    print(x**2)
```

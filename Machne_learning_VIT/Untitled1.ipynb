* Logistic Regression is used when the dependent variable(target) is categorical.
**Example:**
* To predict whether an email is spam (1) or (0)
* Whether the tumor is malignant (1) or not (0)
### Why use logistic resgression and not Linear regression
* Linear regression is not suitable for classification problem.
* Linear regression is unbounded, and this brings logistic regression into picture.
* Their value strictly ranges from 0 to 1.

If we use linear regression for the problem of classifing malignant vs non-malignant problem, there is a **need for setting up a threshold** based on which **classification** can be done. 
Say if the actual class is malignant, predicted continuous value 0.4 and the threshold value is 0.5, the data point will be classified as not malignant which can lead to serious consequence in real time.

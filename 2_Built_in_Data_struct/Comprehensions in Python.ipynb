### Comprehensions in Python
* Provide us with a **short and concise way to construct new sequences** (such as lists, set, dictionary etc.) using sequences which have been already defined.
#### 4 types of comprehensions:

**List Comprehensions**
<br>
* output_list = [output_exp for var in input_list if (var satisfies this condition)]
* list comprehension may or may not contain an if condition. List comprehensions can contain multiple for (nested list comprehensions).
* **Example:** create an output list which contains only the even numbers which are present in the input list.

* **Example:** create an output list which contains squares of all the numbers from 1 to 9.


**Dictionary Comprehensions**
<br>
* output_dict = {key:value for (key, value) in iterable if (key, value satisfy this condition)}
* **Example:** create an output dictionary which contains only the odd numbers that are present in the input list as keys and their cubes as values.

* **Example:** Given two lists containing the names of states and their corresponding capitals, construct a dictionary which maps the states with their respective capitals.



**Set Comprehensions**
<br>
* **Example:** create an output set which contains only the even numbers that are present in the input list. 


**Generator Comprehensions**
<br>
*  Similar to list comprehensions. One difference between them is that generator comprehensions use circular brackets whereas list comprehensions use square brackets.
* major difference between them is that generators don’t allocate memory for the whole list. Instead, they generate each value one by one which is why they are memory efficient.
